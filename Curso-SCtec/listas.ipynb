# Listas

Notebook para estudos e exercicios com listas, generators e dicionarios em Python.

```python
from typing import Any, Generator


dados: list[int] = [1, 2, 3, 4, 5]
soma: Generator[int, None, None] = (soma * 2 for soma in dados)
list(soma)
```

```python
meninos: dict[str, Any] = {
    "nome": "Miguel",
    "sobrenome": "magro",
    "idade": 25,
}

meninas: dict[str, Any] = {
    "nome": ["Maria", "Luiza"],
    "sobrenome": ["Silva", "Alvaras"],
    "idade": [30, 28],
}

print(meninos["nome"])
print(meninos["sobrenome"])
print(meninos["idade"])

print(meninas["nome"])
```
